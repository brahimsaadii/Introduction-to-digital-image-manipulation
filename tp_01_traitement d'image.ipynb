<h2 style='color:red'>Qst 01</h2>

<h3>1 creation of matrixes A, B and C</h3>

```python
i = 0
A = [0]
while A[i]<252:
    i  = i+1
    A.append(A[i-1] + 4)

A = np.array(A)
B = np.zeros(8)
C = np.array([0,0,1,1,1,1,0,0])
```

<h2>cretion des matrice I1, I2 et I3</h2>

<h3>I1 ayant 64 lignes identiques au vecteur A</h3>

```python
I1 = []
for i in A:
    I1.append(A)
I1 = np.array(I1)
I1.shape
```

<h3>I2 constitué d’un bloc de 4X4 de ‘1’ entouré de deux rangs de 0</h3>

```python
I2 = []
for i in range(8):
    if(C[i]==0):
        I2.append(B)
    else:
        I2.append(C)
I2 = np.array(I2)
```

I3 une matrice carrée obtenue par une reproduction 2x2 de la matrice I2

```python
I3 = np.concatenate((I2,I2), axis=1)               
I3 = np.concatenate((I3,I3), axis=0)
I3.shape
```

<h2  style='color:red'>Qst 02</h2>

<h3>convertisent des matrices au type uint8</h3>

```python
I1u = np.uint8(I1)
I2u = np.uint8(I2)
I3u = np.uint8(I3)
```

<b>visualisation des matrices</b>

<b>on utilisant la fonction imshow() du cv2</b>

```python
cv.imshow('img', I1u)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imshow('img', I2u)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imshow('img', I3u)
cv.waitKey(0)
cv.destroyAllWindows()
```

<b>on utilisant la fonction du matplotlib</b>

```python
plt.figure(figsize=(20,10))

plt.subplot(1,3,1)
plt.imshow(I1u, cmap = 'gray')

plt.subplot(1,3,2)
plt.imshow(I2u, cmap = 'gray')

plt.subplot(1,3,3)
plt.imshow(I3u, cmap = 'gray')
```

```python
x, y = 28, 4
IMG = np.zeros([x,y*7,3])
array = np.array((0,1,2,3))
for i in range(x):
    for j in range(y*7):
        for k in range(3):
            if((k == 2) and (j in array)):
                IMG[i,j,k] = int(255)
                
            if((k == 1) and (j in (array + 4))):
                IMG[i,j,k] = int(255)
                
            if((k == 0) and (j in (array + 8))):
                IMG[i,j,k] = int(255)
                
            if((k == 1 or k==0 or k==2) and (j in (array + 12))):
                IMG[i,j,k] = int(255)
            
            if((k == 1  or k==2) and (j in (array + 16))):
                IMG[i,j,k] = int(255)
                
            if((k == 1 or k==0) and (j in (array + 20))):
                IMG[i,j,k] = int(255)    
            
            if((k == 2 or k==0) and (j in (array + 24))):
                IMG[i,j,k] = int(255) 
                
                
                
IMG = np.uint8(IMG)
# IMG = IMG.astype("int")              
plt.imshow(IMG)
plt.show()
```

<h2 style='color:red'>Qst 03</h2>

<b>Pourquoi les valeurs de l’image sont de type « uint8 » ?</b>

L'image est en couleur (RGB) et chaque couleur a 256 niveaux = 2^8 donc codé sur 8 bits en binaire

<h2 style='color:red'>Qst 04</h2>

Separation des canaux RGB du mandrill

```python
R, G, B = cv.split(img_orig)
plt.figure(figsize=(20,10))

plt.subplot(1,3,1)
plt.imshow(R,cmap="gray")
plt.title("ROUGE")

plt.subplot(1,3,2)
plt.imshow(G,cmap="gray")
plt.title("VERT")

plt.subplot(1,3,3)
plt.imshow(B, cmap="gray")
plt.title("BLEU")
```

```python
#initiallisation des variables
RBG = img_orig.copy()
BRG = img_orig.copy()
GRB = img_orig.copy()
GBR = img_orig.copy()

#bgr
BGR = cv.imread("images/mandrill.png")     #la fonction cv.imread lit l'image en format BGR

#rbg
RBG[:,:,0] = R
RBG[:,:,1] = B
RBG[:,:,2] = G

#brg
BRG[:,:,0] = B
BRG[:,:,1] = R
BRG[:,:,2] = G

#grb
GRB[:,:,0] = G
GRB[:,:,1] = R
GRB[:,:,2] = B

#gbr
GRB[:,:,0] = G
GRB[:,:,1] = B
GRB[:,:,2] = R
```
